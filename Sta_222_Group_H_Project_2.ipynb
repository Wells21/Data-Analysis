**Question 1(vi)**

Calculate the mean, median and standard deviation Slope for each of Vegetation
type using aggregate function.

```r
# To get the mean, median and standard deviation on Slope for each vegetation type, we use the function "aggregate"
agg_mean <- aggregate(Slope~Vegetation, data = sta222, FUN = mean)
colnames(agg_mean) <- c("Vegetation", "Slope Mean")
agg_mean
```

For the above dataframe, we calculated the mean values of Slope accross all the Vegation types using the function (aggregate). 

This symbol ~ seperates the dependent variable form the independent variable.

The parameter "FUN" specifies the aggregate function to use, in this case, the mean function.

We can see that the Vegetation type with the highest mean value is **"Arable"** and the lowest mean value is **"Orchard"**

```r
# getting the median value of Slope accross the Vegetation types
agg_median <- aggregate(Slope~Vegetation, data = sta222, FUN = median)
colnames(agg_median) <- c("Vegetation", "Slope Median")
agg_median
```

The dataframe consists of the median values of Slope accross the Vegetation types

```r
# getting the standard deviation value of Slope accross the Vegetation types
agg_sd <- aggregate(Slope~Vegetation, data = sta222, FUN = sd)
colnames(agg_sd) <- c("Vegetation", "Slope SD")
agg_sd
```

A standard deviation (or ) is a measure of how dispersed the data is in relation to the mean. Low standard deviation means data are clustered around the mean, and high standard deviation indicates data are more spread out.

On knowing that, we can then say that The Vegetation type which has more data spread out in the dataframe is **Grassland**, while the Vegetation type which has data clustered around the mean is **Orchard**.

I would start the solution to the above problem by explaining some terms.

Linear regression is a regression model that uses a straight line to describe the relationship between variables. It finds the line of best fit through your data by searching for the value of the regression coefficient(s) that minimizes the total error of the model.

A typical model takes the term **response ~ predictors**, where the response is (Area) and the predictors are (Worm.density and Slope)

Y = a + b1X1 + b2X2

 is a multiple linear regression model that deals with 2 independent variables and 1 dependent variable, where **a** is the intercept and **b1** and **b2** are the coefficients . 

 For our Group (Group H), our 2 independent variables are **(Worm.density** and **Slope)** while our dependent variable is the **Area**

 Therefore, we would be fitting a regression model in the form:

 **Area = a + b1(Worm.density) + b2(Slope)**

 which predicts **Area** from the predictors (**Worm.density** and **Slope**)

```r
# fitting a linear regression model using the function lm()
linear_model <- lm(Area ~ Worm.density + Slope, data = sta222)
```

```r
# getting the information of the model, we use the summary() function
summary(linear_model)
```

***Interpretation of the effect of Worm.density in the model***.

The model suggests a positive association between the Worm.density and the Overall Area; a one-unit increase on the variable Worm.density predicts an increase of 0.08 units on the overall Area (se = 0.097); this increase is slightly significant, t value = 0.8.

***Interpretation of the effect of Slope in the model***.

The model suggests a Negative association between the Slope and the Overall Area; a one-unit increase on the variable Slope predicts a Decrease of -0.05 units on the overall Area (se = 0.07); t value = 0.8.

***Model fit statistics and model significance***

The **residual standard error** is used to measure how well a regression model fits a dataset.

The smaller the **residual standard error**, the better a regression model fits a dataset. Conversely, the higher the residual standard error, the worse a regression model fits a dataset.

The residuals of this model is (the difference between the observed values and the predicted values) also called the error

Our model's **residual standard error** = 1.097

This means that our model predicts the Area with an average error of about 1.097.



```r
# here is the plot of the model which uses a function plot()
plot(linear_model)
```
